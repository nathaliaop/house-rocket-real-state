```python
import pandas as pd
from scipy import stats
# import streamlit as st
import numpy as np
import matplotlib
import seaborn as sns

# new price and old price
def percentage_change(col2,col1):
    return ((col2 - col1) / col1) * 100

pd.set_option('display.float_format', lambda x: '%.2f' % x)

# read dataset
df = pd.read_csv('datasets/kc_house_data.csv')

# ===================
# Tratamento de dados
# ===================

# tipos de dados
df['date'] = pd.to_datetime(df['date'])
# df['yr_built'] = pd.to_datetime(df['yr_built']).year
# df['yr_renovated'] = pd.to_datetime(df['yr_renovated']).year
df['bathrooms'] = df['bathrooms'].astype(int)
df['floors'] = df['floors'].astype(int)
df['waterfront'] = df['waterfront'].apply(lambda x: 'yes' if x == 1 else 'no')
df['m2_basement'] = df['sqft_basement']*0.3048
df['m2_living'] =  df['sqft_living']*0.3048
df['m2_lot'] = df['sqft_lot']*0.3048

# seleciona só as colunas necessárias
df = df[['id', 'date', 'price', 'bedrooms', 'bathrooms', 'm2_living',
       'm2_lot', 'floors', 'waterfront', 'condition',
       'm2_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']]

# remove outliers with trimmed mean 10%
stats.trim_mean(df.select_dtypes(include=np.number), 0.1)

# functions
# get data
# clean data
# buy and sell data

# ==========
# Hypotese 1 - Imóveis que possuem vista para água, são 30% mais caros, na média.
# ==========

mean_waterfront_yes = df.loc[df['waterfront'] == 'yes', 'price'].mean()

mean_waterfront_no = df.loc[df['waterfront'] == 'no', 'price'].mean()

print(percentage_change(mean_waterfront_yes, mean_waterfront_no))



import seaborn as sns

# violin plot (learn to customize)
# ax = sns.violinplot(x=df['waterfront'], y=df['price'], palette="Greens", inner='quartile')
# ax.set_xlabel('Has view to water')

# ax = sns.barplot(x='waterfront', y='price', data=df)
# ax.set_xlabel('Has view to water')

ax = sns.lineplot(x = 'yr_built', y = 'price', data = df)
```
